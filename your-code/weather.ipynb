## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
```

```python
weather = pd.read_csv('../weather-raw.csv')
weather
```

```python
weather.isnull().sum()
#shows that some columns are empty, therefore they can be deleted to make the dataset not unnecessarily longer
```

```python
# weather.drop(['d9','d12','d18','d19','d20','d21','d22','d24'], axis=1)
weather = weather.dropna(axis=1,how='all')   #removes all columns that don't contain a value
weather
```

```python
weather = weather.melt(id_vars=['id','year','month','element'],
                 value_vars=['d1','d2','d3','d4','d4','d6','d7','d8','d10','d11','d13','d14','d15','d16','d17','d23','d25','d26','d27','d28','d29','d30','d31'],
                 var_name='day',
                 value_name='temperature'
                )
weather
```

```python
weather['day'].dtypes
```

```python
weather['day'] = weather['day'].str.replace('d', '')
weather
```

```python
weather.isnull().sum()
```

```python
weather = weather.dropna(axis=0,how='any')
weather.head()
```

```python
weather = weather.pivot(index=['id','year','month','day'],
                        columns=['element'],
                        values=['temperature']
                        )
weather
```
